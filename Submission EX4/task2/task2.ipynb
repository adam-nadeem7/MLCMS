```python

# Define two functions representing stable and unstable states

def eq7_1(r):
    return sqrt(r - 3) / sqrt(2.0)

def eq7_2(r):
    return -sqrt(r - 3) / sqrt(2.0)
# Generate a domain of values for parameter α

domain = linspace(3, 5)
# Plot the saddle-node bifurcation diagram


plot_bif_2d(eq7_1,eq7_2,domain)
```

```python


# Define the differential equations for task2_6 and task2_7
alpha = 1
task2_6 = lambda y: alpha - y**2
task2_7 = lambda y: alpha - 2*y**2 - 3

def plot_trajectory_1d(task, y0, t, title):
    """
    Plot the one-dimensional trajectory using Euler's method for a given differential equation.
    :param task: Function representing the differential equation.
    :param y0: Initial condition.
    :param t: Total time for simulation.
    :param title: Title for the plot.
    """
    # Define the time array
    time = np.linspace(0, t, 100)

    # Solve the differential equation using Euler's method
    yt, time = solve_euler(task, y0, time)

    # Plot the one-dimensional trajectory in red
    plt.plot(time, yt, c='red')

    # Mark the initial point with a green dot
    plt.scatter(0, y0[0], color='green', marker='o')
    plt.title(title + ': α=' + str(alpha) + ', y0=' + str(y0[0]))
    plt.savefig(title + ': α=' + str(alpha) + ', y0=' + str(y0[0]) + '.png')
    plt.show()

def plot_with_y0_list():
    """
    Plot trajectories for multiple initial conditions and differential equations.
    """
    for y0 in y0_list:
        y0 = np.array([y0])
        plot_trajectory_1d(task2_6, y0, 10, 'task2_6') 
        plot_trajectory_1d(task2_7, y0, 10, 'task2_7') 

# Set the initial conditions for alpha = 1
alpha = 1
y0_list = [0.5, 1.5]
plot_with_y0_list()

# Set the initial conditions for alpha = -1
alpha = -1
y0_list = [-0.5, -1.5]
plot_with_y0_list()
```
